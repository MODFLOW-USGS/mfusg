```python
import flopy
```

fname = 'clntest.oc'
with open(fname, 'w') as f:
    
    f.write('HEAD SAVE UNIT 30' + '\n')
    f.write('COMPACT BUDGET' + '\n')
    f.write('\n')
    
    for i in range(30):
        f.write('PERIOD 1 STEP {}\n'.format(i + 1))
        f.write('  PRINT BUDGET' + '\n')
        f.write('  SAVE BUDGET' + '\n')
        f.write('  PRINT HEAD' + '\n')
        f.write('  SAVE HEAD' + '\n')
        f.write('\n')

```python
fname = 'clntest.cln.cbc'
budobj = flopy.utils.binaryfile.CellBudgetFile(fname)
print(budobj.recordarray.dtype.names)
budobj.list_records()
```

```python
for btxt in budobj.get_unique_record_names():
    txt = btxt.decode()
    print(txt,budobj.get_data(text=txt))
```
