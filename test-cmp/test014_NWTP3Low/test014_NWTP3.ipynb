# Reproduction of MODFLOW-NWT Problem 3



## MODFLOW-NWT

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
import flopy
import flopy.utils.binaryfile as bf
```

```python
#paths to existing output
mfnwtpth = os.path.join('mfnwt.cmp')
```

### Get base model

```python
model = 'NWTP3.nam'
ml = flopy.modflow.Modflow.load(model, version='mfnwt', model_ws=mfnwtpth, verbose=True)
```

### Low recharge rate

```python
#--model info
nrow, ncol = ml.dis.nrow, ml.dis.ncol
aqbottom = ml.dis.botm.array[0, :, :]

#--Create the headfile objects
#--MODFLOW-NWT
headfile = os.path.join(mfnwtpth, 'NWTP3.hds')
nwtobj = bf.HeadFile(headfile)
times = nwtobj.get_times()
head = nwtobj.get_data(totim=times[-1]) 
nwthead = head[-1, :, :]
nwtsat = nwthead - aqbottom[:, :]
idx = nwthead < aqbottom + 0.02
nwthead[idx] = -999.
nwtsat[idx] = 0.

            
#--MODFLOW-USG
headfile = os.path.join('nwtp3.hds')
mfusgobj = bf.HeadFile(headfile)
times = nwtobj.get_times()
head = mfusgobj.get_data(totim=times[-1]) 
mfusghead = head[-1, :, :]
mfusgsat = mfusghead - aqbottom[:, :]
idx = mfusghead < aqbottom + 0.02
mfusghead[idx] = -999.
mfusgsat[idx] = 0.

idx = (nwthead == -999.) | (mfusghead == -999.)
nwthead[idx] = -999.
mfusghead[idx] = -999.
nwtsat[idx] = 0.
mfusgsat[idx] = 0.
```

### Simulated head

```python
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(1, 3, 1)
ax.set_aspect('equal')
mm = flopy.plot.ModelMap(model=ml, ax=ax)
h = mm.plot_array(nwthead, cmap='jet', vmin=20, vmax=80, masked_values=[-999.])
c = mm.contour_array(nwthead, colors='k', levels=np.arange(0,100,5))
ax.clabel(c, fmt='%1.0f')
plt.colorbar(h, shrink=0.5)
plt.title('MODFLOW-NWT', size=8)

ax = fig.add_subplot(1, 3, 2)
ax.set_aspect('equal')
mm = flopy.plot.ModelMap(model=ml, ax=ax)
h = mm.plot_array(mfusghead, cmap='jet', vmin=20, vmax=80, masked_values=[-999.])
c = mm.contour_array(mfusghead, colors='k', levels=np.arange(0,100,5))
ax.clabel(c, fmt='%1.0f')
plt.colorbar(h, shrink=0.5)
plt.title('MODFLOW-USG', size=8)

diff = mfusghead - nwthead
ax = fig.add_subplot(1, 3, 3)
ax.set_aspect('equal')
mm = flopy.plot.ModelMap(model=ml, ax=ax)
h = mm.plot_array(diff, cmap='jet', masked_values=[0.])
c = mm.contour_array(diff, colors='k')
ax.clabel(c, fmt='%3.2f')
plt.colorbar(h, shrink=0.5)
plt.title('Difference', size=8);
```

### Saturated thickness

```python
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(1, 3, 1)
ax.set_aspect('equal')
mm = flopy.plot.ModelMap(model=ml, ax=ax)
h = mm.plot_array(nwtsat, cmap='jet', vmin=0, vmax=30, masked_values=[0.])
c = mm.contour_array(nwtsat, colors='k', levels=np.arange(0,100,5))
ax.clabel(c, fmt='%1.0f')
plt.colorbar(h, shrink=0.5)
plt.title('MODFLOW-NWT', size=8)

ax = fig.add_subplot(1, 3, 2)
ax.set_aspect('equal')
mm = flopy.plot.ModelMap(model=ml, ax=ax)
h = mm.plot_array(mfusgsat, cmap='jet', vmin=0, vmax=30, masked_values=[0.])
c = mm.contour_array(mfusgsat, colors='k', levels=np.arange(0,100,5))
ax.clabel(c, fmt='%1.0f')
plt.colorbar(h, shrink=0.5)
plt.title('MODFLOW-USG', size=8)

diff = mfusgsat - nwtsat
ax = fig.add_subplot(1, 3, 3)
ax.set_aspect('equal')
mm = flopy.plot.ModelMap(model=ml, ax=ax)
h = mm.plot_array(diff, cmap='jet', masked_values=[0.])
c = mm.contour_array(diff, colors='k')
ax.clabel(c, fmt='%3.2f')
plt.colorbar(h, shrink=0.5)
plt.title('Difference', size=8);
```
